```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/mbti-type/mbti_1.csv')
df.head()
```

```python
df['words/comment'] = df['posts'].apply(lambda x: len(x.split())/50)
df
```

```python
plt.figure(figsize=(24,16))
sns.boxplot(x='type', y='words/comment', data=df, palette='Set1')
sns.swarmplot(x='type', y='words/comment', data=df, alpha=0.75,palette='Set3').set_title('words per comment for each personality')
```

```python
#seperate the upper figure into 2 figures(1)
plt.figure(figsize=(24,16))
sns.boxplot(x='type', y='words/comment', data=df, palette='Set1').set_title('words per comment for each personality')
```

```python
#seperate the upper figure into 2 figures(2)
plt.figure(figsize=(24,16))
sns.swarmplot(x='type', y='words/comment', data=df).set_title('words per comment for each personality')
```

```python
plt.figure(figsize=(20,10))
plt.ylabel('Number of Occurrences', fontsize=16)
plt.xlabel('Personality Types', fontsize=16)
plt.xticks(fontsize=16, rotation=0)
plt.yticks(fontsize=16, rotation=0)
sns.countplot(data=df, x='type')
#sns.countplot(cnt_srs.index, cnt_srs.values)
```

```python
df = pd.read_csv('/kaggle/input/mbti-type/mbti_1.csv')
# train = pd.read_csv('../input/mbti-type/mbti_1.csv')
mbti = {'I':'Introversion', 'E':'Extroversion', 'N':'Intuition', 
        'S':'Sensing', 'T':'Thinking', 'F': 'Feeling', 
        'J':'Judging', 'P': 'Perceiving'}
cnt_srs = df['type'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(cnt_srs.index, cnt_srs.values)
plt.ylabel('count', fontsize=16)
plt.xlabel('type', fontsize=16)
plt.xticks(fontsize=16, rotation=0)
plt.yticks(fontsize=16, rotation=0)
plt.title('Number of subjuects for each personality')
plt.show()
```

```python
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.offline as py

trace = go.Pie(labels=cnt_srs.index, values=cnt_srs.values)
layout = go.Layout(title='Personality Distribution')
data = [trace]
fig = go.Figure(data = data, layout = layout)
py.iplot(fig)
```
